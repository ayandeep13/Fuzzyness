```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Trapezoidal Membership Function
def trapezoidal(x, a, b, c, d):
    return np.piecewise(x,
                        [x <= a, (x > a) & (x <= b), (x > b) & (x <= c), (x > c) & (x <= d), x > d],
                        [0, lambda x: (x - a) / (b - a), 1, lambda x: (d - x) / (d - c), 0])
```

```python
# 2. Triangular Membership Function
def triangular(x, a, b, c):
    return np.piecewise(x,
                        [x <= a, (x > a) & (x <= b), (x > b) & (x <= c), x > c],
                        [0, lambda x: (x - a) / (b - a), lambda x: (c - x) / (c - b), 0])
```

```python
# 3. Gaussian Membership Function
def gaussian(x, c, sigma):
    return np.exp(-((x - c)**2) / (2 * sigma**2))
```

```python
# Main function to plot the fuzzy numbers
def plot_fuzzy_numbers():
    # Generate x values
    x = np.linspace(0, 8, 1000)

    # Parameters for the membership functions
    a_trap, b_trap, c_trap, d_trap = 2, 3, 5, 6
    a_tri, b_tri, c_tri = 2, 4, 6
    c_gauss, sigma_gauss = 4, 1

    # Calculating membership values
    trap_values = trapezoidal(x, a_trap, b_trap, c_trap, d_trap)
    tri_values = triangular(x, a_tri, b_tri, c_tri)
    gauss_values = gaussian(x, c_gauss, sigma_gauss)

    # Plotting the membership functions
    plt.figure(figsize=(10, 6))

    plt.plot(x, trap_values, label="Trapezoidal", color='blue')
    plt.plot(x, tri_values, label="Triangular", color='green')
    plt.plot(x, gauss_values, label="Gaussian", color='red')

    plt.title("Fuzzy Numbers 'Approximately Equal to 4'")
    plt.xlabel("x")
    plt.ylabel("Membership Degree")
    plt.legend()
    plt.grid(True)
    plt.show()

# Calling the function to plot the fuzzy numbers
plot_fuzzy_numbers()
```
