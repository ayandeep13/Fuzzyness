```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def memberA(x):
    if x < 30:
        return 1
    elif 30 <= x < 50:
        return (50 - x) / (50 - 30)
    elif 50 <= x < 70:
        return (70 - x) / (70 - 50)
    else:
        return 0

def memberB(x):
    if x < 60:
        return 0
    elif 60 <= x < 80:
        return (x - 60) / (80 - 60)
    elif 80 <= x < 100:
        return (x - 80) / (100 - 80)
    else:
        return 1
```

```python
Ax1 = np.linspace(0, 30, 100)
Ax2 = np.linspace(30, 50, 100)
Ax3 = np.linspace(50, 70, 100)
Ax4 = np.linspace(70, 120, 100)
Bx1 = np.linspace(0, 60, 100)
Bx2 = np.linspace(60, 80, 100)
Bx3 = np.linspace(80, 100, 100)
Bx4 = np.linspace(100, 120, 100)
```

```python
# Range of x values (from 0 to 120 for better visualization)
x_values = np.linspace(0, 120, 1000)

# Membership values for A(x) and B(x)
memberA_values = np.array([memberA(x) for x in x_values])
memberB_values = np.array([memberB(x) for x in x_values])

# Union (A ∪ B)
memberunion = np.maximum(memberA_values, memberB_values)

# Intersection (A ∩ B)
memberintersection = np.minimum(memberA_values, memberB_values)

# Complement (A^c and B^c)
memberA_complement = 1 - memberA_values
memberB_complement = 1 - memberB_values

# Difference (A - B)
memberdifference = memberA_values - memberB_values
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)
plt.plot(x_values, memberA_values, label='A(x)', color = 'Blue')
plt.title('A')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()


plt.subplot(2, 2, 2)
plt.plot(x_values, memberB_values, label='B(x)', color = 'Orange')
plt.title('B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
```

```python
plt.figure(figsize=(14, 10))

# Plotting for Union
plt.subplot(2, 2, 1)
plt.plot(x_values, memberA_values, label='A(x)')
plt.plot(x_values, memberB_values, label='B(x)')
plt.plot(x_values, memberunion, label='A U B', linestyle='--',linewidth=3)
plt.title('Union (A U B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

# Plotting for Intersection
plt.subplot(2, 2, 2)
plt.plot(x_values, memberA_values, label='A(x)')
plt.plot(x_values, memberB_values, label='B(x)')
plt.plot(x_values, memberintersection, label='A ∩ B', linestyle='--',linewidth=3)
plt.title('Intersection (A ∩ B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

# Plotting for Complements
plt.subplot(2, 2, 3)
plt.plot(x_values, memberA_complement, label="A' (complement of A)")
plt.plot(x_values, memberB_complement, label="B' (complement of B)")
plt.title('Complements (A^c and B^c)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

# Plotting for Difference
plt.subplot(2, 2, 4)
plt.plot(x_values, memberA_values, label='A(x)')
plt.plot(x_values, memberB_values, label='B(x)')
plt.plot(x_values, memberdifference, label='A - B', linestyle='--',linewidth=3)
plt.title('Difference (A - B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```
